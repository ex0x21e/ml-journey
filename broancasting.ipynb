# Введение в бродкастинг

В этом примере мы подробно изучаем, как бродкастинг (broadcasting) позволяет выполнять арифметические операции между массивами разной размерности.

## Что такое бродкастинг?

Бродкастинг — это набор правил, по которым `numpy` автоматически расширяет меньший массив, чтобы его форма согласовалась с формой большего массива.

Правила бродкастинга просты:
1. Если массивы имеют разное число измерений, к массиву с меньшей размерностью добавляются единицы слева.
2. Два измерения считаются совместимыми, если они равны или один из них равен 1.
3. Если размерность равна 1, массив трактуется как повторяемый вдоль этого измерения.

## Исходные данные

### Цены смартфонов

| Smartphone | Price |
| --- | ---: |
| Iphone | 300 |
| Xiaomi | 150 |
| Samsung | 200 |

### Стоимость доставки

| Delivery type | Cost |
| --- | ---: |
| standard | 5 |
| express | 10 |
| avia_sales | 20 |

## Формы массивов

| Массив | Форма | Описание |
| --- | --- | --- |
| `w` | `(3, 1)` | столбцовый вектор цен смартфонов |
| `v` | `(1, 3)` | строковый вектор стоимости доставки |
| `result` | `(3, 3)` | матрица, в которой каждая цена сочетается с каждой доставкой |

## Почему это полезно?

Бродкастинг избавляет от необходимости писать вложенные циклы или вручную дублировать данные. Он делает код компактным, эффективным и похожим на матричные операции в линейной алгебре.

## Что делает код дальше?

1. Создает `DataFrame` с ценами смартфонов и `Series` со стоимостью доставки.
2. Переводит их в `numpy`-массивы и изменяет форму с помощью `reshape`.
3. Складывает массивы, используя бродкастинг.
4. Возвращает результат обратно в `DataFrame` для удобного отображения.

## Ожидаемый результат

| Smartphone | standard | express | avia_sales |
| --- | ---: | ---: | ---: |
| Iphone | 305 | 310 | 320 |
| Xiaomi | 155 | 160 | 170 |
| Samsung | 205 | 210 | 220 |

```python

import pandas as pd

delivery = {
    "standard": 5,
    "express": 10,
    "avia_sales": 20,
}

data: dict[str, list[str] | list[int]] = {
    "Smartphone": ["Iphone", "Xiaomi", "Samsung"],
    "Price": [300, 150, 200]
}

df_smartphones = pd.DataFrame(data).set_index("Smartphone")  # индексы строк — названия смартфонов
df_delivery = pd.Series(delivery)  # серия со стоимостью доставки по типам

w = df_smartphones["Price"].to_numpy().reshape(-1, 1)  # столбцовый вектор shape (3, 1) -1 начит что кол-во строк вычисляется автоматически
v = df_delivery.to_numpy().reshape(1, -1)  # строковый вектор shape (1, 3)

result = w + v

df_final = pd.DataFrame(result,
                        index=df_smartphones.index,
                        columns=df_delivery.index,
                        )

df_final
```
